# Simple model

As data describes number ($y_i$) of deaths due to airplane accidents in one year (for 10 consequent years), assuming that data can has Poisson distribution seems to reasonable, so we denote: </br>
$y_i\sim Poisson(\lambda_i)$

We assume that the death rate (described using $\lambda_i$ parameter) is constant throughout all the years, and is independent from other variables we denote: </br>
$\lambda_i=\widetilde{\lambda}$

### Prior modelling

From common knowledge, airplanes are safest means of travel.
Airplane accidents are very rare, but usually a lot of people die in a single one.
It's safe to assume that accidents happen less often than once every two days.

Biggest passenger airplane that was used in '80-s was Boeing 747-300 and it could carry up to 584 people.

Fully seated 747 crashing every second day for a year is very unlikely to happen to occur, so death rate's upper bound should fulfill: </br>
$\widetilde\lambda + 3\sqrt{\widetilde{\lambda}} \approx 183\cdot584$

```python
lambdas = np.polynomial.polynomial.polyroots([-183.*584,3.,1.])
lambda_upper_bound = np.round(lambdas[lambdas>0]**2)[0]
print(lambda_upper_bound)
```

We can assume that $\widetilde\lambda$ has half-normal distribution with standard deviation parameter $\sigma$ that ensures that $99.99\%$  of the probability lies to the left of found upper bound of $\widetilde\lambda$. </br>

```python
with open('stans/simple_prior_tune.stan', 'r') as file:
    print(file.read())
```

```python
simple_prior_tune = stan_utility.compile_model('stans/simple_prior_tune.stan')
```

```python
simple_prior_tuned = simple_prior_tune.sampling(data={'y_guess':np.array([np.log(70000)]),
                                                      'theta':np.array([lambda_upper_bound])},
                                                seed=7044,
                                                algorithm="Fixed_param",
                                                iter=1,
                                                warmup=0,
                                                chains=1)
sigma = np.round(simple_prior_tuned.extract()['sigma'][0])
print(sigma)
```

```python
fig, ax2 = plt.subplots(1, 1,figsize=(7, 4))
x2=np.linspace(0,3*sigma,1000)
x4=np.linspace(0,lambda_upper_bound,1000)
ax2.plot(x2,2*stats.norm.pdf(x2,scale=sigma),color=dark,linewidth=2)
ax2.fill_between(x4,2*stats.norm.pdf(x4,scale=sigma),0,color=dark)
ax2.set_yticks([])
ax2.set_xticks([0,lambda_upper_bound])
ax2.set_title(r'$\lambda$')
plt.show()
```

# More complicated model

Intuitively, number of accidents in a year is dependent on how much flights there are. Of course the more we fly, the more accidents there are, and more people die.

We assume that the death rate (described using $\lambda_i$ parameter) is proportional to number of miles flown ($n_i$) and is independent from other variables: </br>
$y_i=\mathrm{Poisson}(\theta\cdot n_i)$ </br>
$\theta$ is intesity parameter

### Prior modelling

Again we assume that fully seated 747 crashing every second day for a year is very unlikely to happen to occur, so death rate's upper bound should fulfill: </br>
$\theta\bar n + 3\sqrt{\theta\bar n} \approx 183\cdot584$ </br>
$u = \theta\bar n$ </br>
$u + 3\sqrt{u} \approx 183\cdot584$ </br>

```python
u = np.polynomial.polynomial.polyroots([-183.*584,3.,1.])
u = (u[u>0]**2)[0]
theta_ub = u/(data['miles_flown'].mean())
print(theta_ub)
```

We assume that $\theta$ has half-normal distribution and that $99.99%$ of the probability lies to the left of upper bound of $\theta$. </br>
Previously written Stan program can be used here again.

```python
better_prior_tuned = simple_prior_tune.sampling(data={'y_guess':np.array([np.log(30)]),
                                                      'theta':np.array([theta_ub])},
                                                seed=7044,
                                                algorithm="Fixed_param",
                                                iter=1,
                                                warmup=0,
                                                chains=1)
sigma = better_prior_tuned.extract()['sigma'][0]
print(sigma)
```

```python
fig, ax2 = plt.subplots(1, 1,figsize=(7, 4))
x2=np.linspace(0,3*sigma,1000)
x4=np.linspace(0,theta_ub,1000)
ax2.plot(x2,2*stats.norm.pdf(x2,scale=sigma),color=dark,linewidth=2)
ax2.fill_between(x4,2*stats.norm.pdf(x4,scale=sigma),0,color=dark)
ax2.set_yticks([])
ax2.set_xticks([0,sigma,theta_ub])
ax2.set_title(r'$\lambda$')
plt.show()
```
