Columns:
  - accidents - count of airplane accidents in given year
  - deaths - count of deaths caused by the airplane accidents
  - death_rate - average number of deaths for every 100 million miles flown each year
  - miles_flown - sum of all distances flown each year given in x*100 000 000 miles

```python
for c in data.columns:
    data[c].plot(figsize=(7,5))
    plt.ylabel(c)
    plt.show()
```
